## Barchart

## สร้างกราฟแท่งเปรียบเทียบรายได้ในปี 2541 ของจังหวัดกรุงเทพและปริมณฑล

```python
plt.bar([1,2,3,4],greaterbangkok.iloc[:,2]) #เอาทุกแถว,เลือกคอลัมน์ที่ 2
plt.title('Barchart comparing average monthhly income of provinces in Greater Bangkok in 2541 ') 
plt.xlabel('provinces')
plt.ylabel('Income Eath/Month/Household')
```

## Add tick lable  คือ การใส่ชื่อแต่ละแกน

```python
plt.bar([1,2,3,4],greaterbangkok.iloc[:,2],tick_label=greaterbangkok.iloc[:,-1]) 
plt.title('Barchart comparing average monthhly income of provinces in Greater Bangkok in 2541') 
plt.xlabel('provinces')
plt.ylabel('Income Eath/Month/Household')
```

## compare different years คือ การพอตกราฟเปรียบเทียบกัน 2 แท่ง
,align='edge' เพื่อขยับ label ให้มันตรงมากขึ้น

```python
plt.bar([1-(0.35/2),2-(0.35/2),3-(0.35/2),4-(0.35/2)],greaterbangkok.iloc[:,-2],width=0.35,tick_label=greaterbangkok.iloc[:,-1],color ='y',label = '2558',align='edge')
plt.bar([1+(0.35/2),2+(0.35/2),3+(0.35/2),4+(0.35/2)],greaterbangkok.iloc[:,2],width=0.35,tick_label=greaterbangkok.iloc[:,-1],color ='g',label = '2541',align='edge')
plt.title('Barchart comparing average monthhly income of provinces in Greater Bangkok in 2541,2558') 
plt.xlabel('provinces')
plt.legend(loc=1) #บอกว่าสีไหนเป็นปีไหน ถ้า (loc=1): อยู่มุมขวาบน
plt.ylabel('Income Eath/Month/Household')
```

# Radar chart 
คือ กราฟเรดาร์เป็นวิธีที่กราฟิกของการแสดงข้อมูลแบบหลายตัวแปรในรูปแบบของกราฟสองมิติของสามหรือมากกว่าตัวแปรเชิงปริมาณที่แสดงบนแกนที่เริ่มต้นจากจุดเดียวกัน

```python
import pandas as pd
from math import pi
```

```python
pi
```

```python
categories  = [{'client_id': 'c1', 'cheese': 2, 'beer': 2, 'chicken': 5, 'bread': 4, 'coffee':2},
         {'client_id': 'c2', 'cheese': 5, 'beer': 1, 'chicken': 3, 'bread': 3, 'coffee':3},
         {'client_id': 'c3', 'cheese': 4, 'beer': 3, 'chicken': 3, 'bread': 2, 'coffee':5},
         {'client_id': 'c4', 'cheese': 3, 'beer': 5, 'chicken': 4, 'bread': 1, 'coffee':3}]
df = pd.DataFrame(categories)
```

```python
df
```

```python
categories = list(df)[1:]
```

```python
values = df.mean().values.flatten().tolist()
values += values[:1] # repeat the first value to close the circular graph
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8),
                       subplot_kw=dict(polar=True))
plt.xticks(angles[:-1], categories, color='grey', size=12)
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

plt.ylim(0, 5)
ax.set_rlabel_position(30)

ax.plot(angles, values, linewidth=1, linestyle='solid')
ax.fill(angles, values, 'skyblue', alpha=0.4)
```

```python
#prepare canvas
angles = [n / float(len(categories)) * 2 * pi for n in range(len(categories))]
angles += angles[:1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 8),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], categories, color='grey', size=12)
plt.yticks(np.arange(1, 6), ['1', '2', '3', '4', '5'],
           color='grey', size=12)

plt.ylim(0, 5)
ax.set_rlabel_position(30)

# part 1
val_c1 = df.loc[0].drop('client_id').values.flatten().tolist()
val_c1 += val_c1[:1]
ax.plot(angles, val_c1, linewidth=1,
        linestyle='solid', label='Client c1')
ax.fill(angles, val_c1, 'skyblue', alpha=0.4)
 
# part 2
val_c2=df.loc[1].drop('client_id').values.flatten().tolist()
val_c2 += val_c2[:1]
ax.plot(angles, val_c2, linewidth=1,
        linestyle='solid', label='Client c2')
ax.fill(angles, val_c2, 'lightpink', alpha=0.4)
 
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```
