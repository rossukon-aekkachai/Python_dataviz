# Box plot : การกระจายของข้อมูล

### x : Array or a sequence of vectors. คือ ลำดับของตัวเลข

```python
import pandas
import numpy as np
```

```python
irisdata = pandas.read_csv('C:\\Users\\User\\Desktop\\Basic Programming by Aj.Ton\Python_dataviz\\iris.data',header=None)
irisdata
```

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams[ 'figure.figsize'] = [15, 5 ]   #ดูการกระจาย
plt.subplot(1, 4, 1)   # 1 แถว / 4 หลัก / อยู่ตำแหน่งที่ 1
plt.boxplot(setosa_np[:, 0]) 
plt.subplot(1, 4, 2)  
plt.boxplot(setosa_np[:, 1]) 
plt.subplot(1, 4, 3)
plt.boxplot(setosa_np[:, 2]) 
plt.subplot(1, 4, 4)
plt.boxplot(setosa_np[:, 3]) 
plt.show()
```

### use boxplot to compare features between 3 classes : ใช้ boxplot เพื่อเปรียบเทียบคุณสมบัติระหว่าง 3 คลาส

```python
CC = [setosa_np[:,0],versicolor_np[:,0],virginica_np[:,0]]
```

```python
CC
```

```python
plt.boxplot(CC)
plt.title('feature 1 comparison'); #ชื่อของกราฟ   (;)คือ เคลีย taxt output ออกไป
```

```python
# การอธิบายของแต่ละกราฟ  labels = ['setosa','versicolor','virginica'])
```

```python
plt.boxplot(CC,labels = ['setosa','versicolor','virginica'])
plt.title('feature 1 comparison'); #ชื่อของกราฟ   (;)คือ เคลีย taxt output ออกไป
```

## ทำการตั้งชื่อข้อมูลใหม่เพื่อเอาไว้เก็บข้อมูลที่เราทำการลบออกไปแล้ว เหลือแค่ที่เราต้องการพร้อมสำหรับการใช้งาน

```python
Debt_df_debtt =Debt_df [['Region and province',2549]]
```

```python
Debt_df_debtt.drop([0,1,2,3,8,31,49,70,85,86,87,88],axis=0,inplace=True) # ลบแถวที่ไม่เกี่ยวข้องออกไป
```

## ทำการ import folium เพื่อใช้สำหรับการทำแผนที่ โดยใช้คำสั่ง pip install folium

```python
import folium
import pandas as pd
```

```python
data_to_plot = Debt_df_debtt
```

```python
data_to_plot
```

```python
data_to_plot.head()
```

## ทำการ Download ข้อมูลเชิงภูมิศาสตร์ของประเทศไทย 

```python
country_geo = 'thai.json'
```

## และทำการลบช่องว่างข้างหน้าชื่อออกเพื่อที่จะสามารถทำให้ไฟล์ 2 ไฟล์นั้นสามารถเชื่อมกันได้

```python
new_col = list()
for d in data_to_plot['Region and province']:
    new_col.append(d.lstrip())
data_to_plot['Region and province'] = new_col
```

## ทำการกำหนดพิกัด ลองติจูดและละติจูดของประเทศไทย

```python
map = folium.Map(location=[10,100], zoom_start=5)
```

```python
map
```

### การลงสีของข้อมูลที่มีค่าของจำนวนหนี้สินเฉลี่ยต่อครัวเรือนถ้าจังหวัดไหนที่มีจำนวนหนี้สินเฉลี่ยต่อครัวเรือน
     ถ้ามี มีค่ามากจะเป็นสีม่วงเข้ม
     แต่ถ้า มีค่าน้อยจะเป็นสีเหลืองตามแถบรูปข้างบน


```python
map.choropleth(geo_data = country_geo, data=data_to_plot,
              columns=['Region and province',2549],
               key_on='feature.properties.name',
              fill_color='YlGnBu',fill_opacity=0.7,line_opacity=0.2,
              legend_name='Average Debt')      
```

```python
map.save('plot_data.html') #การ save ไฟล์เป็นแบบเว็บไซต์ (html)
```
