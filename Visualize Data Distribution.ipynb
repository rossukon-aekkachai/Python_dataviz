## Scatter plot : ดูว่าข้อมูลมีการกระจายอย่างไร

```python
irisdata = pandas.read_csv('C:\\Users\\User\\Desktop\\Basic Programming by Aj.Ton\Python_dataviz\\iris.data',header=None)
irisdata
```

##### ตัวแรกเป็น x ตัวที่2 เป็น y

##### ตัวเลขตัวแรก บอกแถว(แนวนอน) ตัวที่ 2 บอกคอลัมน์ (หลักแนวตั้ง)

```python
setosa_np = irisdata.iloc[:50,:-1].to_numpy()
setosa_np[:5,:]
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1])
```

##### Scalar ตัวเลข 1 ตัว ซึ่งบอกแต่ขนาด (s คือขนาดของจุด)

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=60)
```

#### แกน X เป็นหลักที่ 1 แกน y เป็นหลักที่ 2 และขนาดของจุด เป็นหลักที่ 3

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=setosa_np[:,2])
```

#### 10* คือ การเพิ่มขนาด

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2])
```

#### การเปลี่ยนสี ให้เป็นพารามิเตอร์ตัวที่ 4

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=10*setosa_np[:,2], c = 'c')
```

```python
print(f'max = {max(setosa_np[:,3])}, min = {min(setosa_np[:,3])}')
```

```python
colors = []
for i in setosa_np[:,3]:
    if i < 0.15:
        colors.append('b')
    elif i<0.3:
        colors.append('g')
    elif i<0.45:
        colors.append('y')
    else:
        colors.append('r')
colors
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = colors)
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='YlOrBr')
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = np.array([[1,0.5,0.8]]))
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='YlOrBr')
```

## Visualize 2 data : ลงไปในกราฟเดียวกัน เพื่อดูความแตกต่าง

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='BuPu',marker='p')
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2], c = versicolor_np[:,3],cmap='BuPu',marker='*')
```

```python
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

```python
plt.scatter(setosa_np[:,0],setosa_np[:,1],s=30*setosa_np[:,2], c = setosa_np[:,3],cmap='BuPu',marker='p')
plt.scatter(versicolor_np[:,0],versicolor_np[:,1],s=30*versicolor_np[:,2], c = versicolor_np[:,3],cmap='BuPu',marker='*')
plt.scatter(virginica_np[:,0],virginica_np[:,1],s=30*virginica_np[:,2], c = virginica_np[:,3],cmap='BuPu',marker='x')
```
