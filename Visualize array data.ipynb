## header = None โหลดไฟล์นี้โดยที่ไม่มีหัวตาราง

```python
irisdata = pandas.read_csv('C:\\Users\\User\\Desktop\\Basic Programming by Aj.Ton\Python_dataviz\\iris.data',header=None)
irisdata
```

```python
import numpy as np
```

#### เราจะสไลด์หรือตัด อยากได้แค่ตัวเลข :แถวของอาเรย์ แนวนอน ,หลักของอาเรย์นั้น แนวตั้ง : ก่อนคอลัมน์ (-1) สุดท้าย

```python
sentosa_np = irisdata.iloc[:50,:-1].to_numpy()
sentosa_np
```

```python
versicolor_np = irisdata.iloc[50:100,:-1].to_numpy()
virginica_np = irisdata.iloc[100:,:-1].to_numpy()
```

## visualize array
    สีสว่าง ค่ามาก สีเข้ม ค่าน้อย

```python
from matplotlib import pyplot as plt
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = [12, 12]
plt.imshow(sentosa_np, interpolation='nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12, 12]
plt.imshow(versicolor_np, interpolation='nearest')
plt.show()
```

```python
matplotlib.rcParams['figure.figsize'] = [12, 12]
plt.imshow(virginica_np, interpolation='nearest')
plt.show()
```

#### plt.subplot คือ plot หลายๆรูปไว้ในรูปเดียว เช่น (1,3,1)  คำสั่งเรียงแถวที่ 1 หลักที่ 3 และเรียงลำดับรูป 1,2,3

```python
matplotlib.rcParams['figure.figsize'] = [12, 12]
plt.subplot(1,3,1)
plt.imshow(sentosa_np, interpolation='nearest')
plt.subplot(1,3,2)
plt.imshow(versicolor_np, interpolation='nearest')
plt.subplot(1,3,3)
plt.imshow(virginica_np, interpolation='nearest')
plt.show()
```
